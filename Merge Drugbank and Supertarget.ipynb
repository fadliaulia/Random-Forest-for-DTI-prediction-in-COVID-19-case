```python
import pandas as pd
import numpy as np
```

```python
drugbank = pd.read_csv('HasilScrapingDrugBank with PUBCHEMID.csv')
supertarget = pd.read_csv('HasilScrapingSuperTarget2.csv')
protein_list = pd.read_csv('GeneCards Protein Only.csv')
```

```python
drugbank_merged = pd.merge(drugbank,protein_list, on='Accession')
drugbank_merged = drugbank_merged.drop(['Actions','Drug group','Category','Gifts','GC ID','Relevance Score','Link', 'Pharmacological action?', 'Details'], axis=1)
drugbank_merged= drugbank_merged.rename(columns={'PuBCHEM ID': 'CID'})
drugbank_merged
```

```python
supertarget_merged = pd.merge(supertarget,protein_list, on='Accession')
supertarget_merged = supertarget_merged.drop(['Category','Gifts','GC ID','Relevance Score','Link'], axis=1)
supertarget_merged
```

```python
drugbank_merged_duplicate = drugbank_merged[drugbank_merged.duplicated(['CID','Accession'])]
print("Duplicate Rows based on 2 columns are:", drugbank_merged_duplicate, sep='\n')
drugbank_clean = drugbank_merged.drop_duplicates(subset=['CID','Accession'])
drugbank_clean = drugbank_clean.rename(columns={'Name': 'Drug Name'})
drugbank_clean 
```

```python
supertarget_merged_duplicate = supertarget_merged[supertarget_merged.duplicated(['CID','Accession'])]
print("Duplicate Rows based on 2 columns are:", supertarget_merged_duplicate, sep='\n')
supertarget_clean = supertarget_merged.drop_duplicates(subset=['CID','Accession'])
supertarget_clean = supertarget_clean.rename(columns={'Drug': 'Drug Name'})
supertarget_clean
```

```python
supertarget_drugbank_merged = pd.concat([supertarget_clean[['CID','Drug Name','Accession','Gene Names']],drugbank_clean[['CID','Drug Name','Accession','Gene Names']]])
supertarget_drugbank_merged
```

```python
supertarget_drugbank_merged_duplicate = supertarget_drugbank_merged[supertarget_drugbank_merged.duplicated(['CID','Accession'])]
print("Duplicate Rows based on 2 columns are:", supertarget_drugbank_merged_duplicate, sep='\n')
clean_data = supertarget_drugbank_merged.drop_duplicates(subset=['CID','Accession'])
clean_data
```

```python
final_df = clean_data.sort_values(by=['Accession'], ascending=False)
final_df = final_df.reset_index(drop=True)
final_df.to_csv('E:\Thesis\Data\SUPERTARGETDRUGBANK_CLEAN_NEW.csv')
```
