```python
import pandas as pd
import numpy as np
import pubchempy as pc
import time
```

```python
dataset = pd.read_csv('SUPERTARGETDRUGBANK_CLEAN_NEW.csv')
dataset
```

```python
list_cid_unik = dataset['CID'].unique()
list_protein_unik = dataset['Accession'].unique()
```

```python
list_protein_unik.shape
```

```python
compound_smiles = []
compound_fingerprint=[]
j=0
for i in list_cid_unik:
    print("senyawa ke", j) 
    print("cid",i)
    t0 = time.time()
    #Smiles
    compound_smiles.append((i,pc.Compound.from_cid(str(i)).canonical_smiles))
    #Fingerprint
    fp = [i]; fp.extend([j for j in pc.Compound.from_cid(str(i)).cactvs_fingerprint])
    compound_fingerprint.append(fp)
    t1 = time.time()
    print("waktu proses", t1-t0)
    j=j+1
com_smiles2 = pd.DataFrame(compound_smiles, columns = ["CID_senyawa", "smiles"])
com_fingerprint1 = pd.DataFrame(compound_fingerprint, columns=["CID_senyawa"] + list(range(1,882)))
```

```python
com_smiles2 = pd.DataFrame(compound_smiles, columns = ["CID_senyawa", "smiles"])
com_fingerprint1.to_csv('com_fingerprint.csv', index=False)
# com_smiles2.to_csv('E:\Thesis\Data\com_smiles2.csv', index=False)
```

```python
import requests
BASE = 'http://www.uniprot.org'
KB_ENDPOINT = '/uniprot/'
TOOL_ENDPOINT = '/uploadlists/'
```

```python
cov_fasta = []
for ids in list_protein_unik:
    payload = {'query': 'mnemonic:{}'.format(ids),
            'format': 'fasta',
            }
    result = requests.get(BASE + KB_ENDPOINT, params=payload)
    if len(result.text) == 0:
        payload['query'] = 'accession:{}'.format(ids.split("_")[0])
        result = requests.get(BASE + KB_ENDPOINT, params=payload)
    cov_fasta.append((ids,result.text))
cov_pro_fasta = pd.DataFrame(cov_fasta, columns = ["pro_uniprot_id", "fasta"])
```

```python
numpy_array=cov_pro_fasta.to_numpy()
np.savetxt("test_file.txt", numpy_array,fmt = "%s")
cov_pro_fasta.to_csv('cov_pro_fasta.csv',index=False)
```

```python
outF = open("cov_fasta.txt", "w+")
for line in cov_pro_fasta["fasta"]:
  # write line to output file
  outF.write(line)
  outF.write("\n")
outF.close()
```
